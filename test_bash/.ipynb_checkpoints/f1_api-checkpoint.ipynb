# Now we need a loop or function to get all the tables

```python
def query_builder(current_year):

    lap = list(range(1,101))*20*(current_year - 1996 + 1)
    rounds = []
    for i in range(1,21):
        rounds.extend(list(itertools.repeat(i,100)))
    rounds = rounds * (current_year - 1996 + 1)

    seasons = []
    for i in range(1996,current_year + 1):
        seasons.extend(list(itertools.repeat(i, 20*100)))

    query_structure_laps = []
    for i in range(0,100*20*(current_year - 1996 + 1)):
        query_structure_laps.append((str(seasons[i]) + '/' + str(rounds[i]) + '/' + 'laps' + '/' + str(lap[i])))

    pit = list(range(1,5))*20*(current_year - 2012 + 1)
    rounds = []
    for i in range(1,21):
        rounds.extend(list(itertools.repeat(i,4)))
    rounds = rounds * (current_year - 2012 + 1) 

    seasons = []
    for i in range(2012, current_year + 1):
        seasons.extend(list(itertools.repeat(i,4*20)))

    query_structure_pits = []
    for i in range(0,4*20*(current_year - 2012 + 1)):
        query_structure_pits.append((str(seasons[i]) + '/' + str(rounds[i]) + '/' + 'pitstops' + '/' + str(pit[i])))
    
    tables = ['seasons','races','results/1','qualifying/1', 'driverStandings/1','constructorStandings/1','drivers'
         ,'constructors','circuits','status']
    
    extended_table = []
    extended_table.extend(query_structure_pits)
    extended_table.extend(query_structure_laps)

    print(tables[0:14])
    print(extended_table[0:9])
    print(extended_table[-9:])
    
query_builder(2019)
```

```python
for table in tables:
    r = requests.get('http://ergast.com/api/f1/{}'.format(table) + '.json')
    if r.status_code != 200:
        print('INVALID QUERY!!! {}'.format(table))
    else:
        print('continuing with {}'.format(table))
        r.json()
    
```
